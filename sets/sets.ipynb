```python
#set5

def gcd(a,b):
    if(b==0):
        return a
    else:
        return gcd(b,a%b)

# print(gcd(6,9))

class Parent:
    def __init__(self, value):
        print("Hello from parent")
        self.value = value
    
    def property(self):
        print(f"Property owned by parent: {self.value}")

class Child(Parent):
    def __init__(self):
        print("Hello from child")
        super().__init__(100)  # Call the parent constructor with value 100


chu = Child()
chu.property()  # Should now correctly access the parent class's property method


class Shape:
    def area_Rectangle(self, length, width):
        return length * width
    def area_Circle(self, radius):
        return 3.14 * radius * radius
    def area_Square(self, side):
        return side * side

myshape=Shape()

print(myshape.area_Rectangle(4,5))  
print(myshape.area_Circle(5))  
print(myshape.area_Square(5))
        
```

```python
#set3

def fib(n):
    if n==0 or n==1:
        return 1
    else:
        return fib(n-1)+fib(n-2)

print(fib(5))

set1={1,2,3,4,5}
set2={4,5,6,7,8}
print(set1.union(set2))
print(set1.intersection(set2))
print(set1.difference(set2))
print(set1.issubset(set2))
print(set1.issuperset(set2))
print(set1.isdisjoint(set2))
print(set1.symmetric_difference(set2))
```
